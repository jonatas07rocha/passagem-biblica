```python
import re
import requests
from bs4 import BeautifulSoup

# Mapeamento de livros da Bíblia com múltiplas variações
bible_books = {
    "genesis": "genesis",
    "gênesis": "genesis",
    "exodo": "exodo",
    "êxodo": "exodo",
    "levitico": "levitico",
    "levítico": "levitico",
    "numeros": "numeros",
    "números": "numeros",
    "deuteronomio": "deuteronomio",
    "deuteronômio": "deuteronomio",
    "josue": "josue",
    "josué": "josue",
    "juizes": "juizes",
    "juízes": "juizes",
    "rute": "rute",
    "1samuel": "isamuel",
    "2samuel": "iisamuel",
    "1reis": "ireis",
    "2reis": "iireis",
    "1cronicas": "icronicas",
    "1crônicas": "icronicas",
    "2cronicas": "iicronicas",
    "2crônicas": "iicronicas",
    "esdras": "esdras",
    "neemias": "neemias",
    "ester": "ester",
    "jó": "jo",
    "salmos": "salmos",
    "salmo": "salmos",
    "proverbios": "proverbios",
    "provérbios": "proverbios",
    "eclesiastes": "eclesiastes",
    "canticos": "canticos",
    "cânticos": "canticos",
    "isaias": "isaias",
    "isaías": "isaias",
    "jeremias": "jeremias",
    "lamentacoes": "lamentacoes",
    "lamentações": "lamentacoes",
    "ezequiel": "ezequiel",
    "daniel": "daniel",
    "oseias": "oseias",
    "oséias": "oseias",
    "joel": "joel",
    "amos": "amos",
    "amós": "amos",
    "obadias": "obadias",
    "jonas": "jonas",
    "miqueias": "miqueias",
    "miquéias": "miqueias",
    "naum": "naum",
    "habacuque": "habacuque",
    "sofonias": "sofonias",
    "ageu": "ageu",
    "zacarias": "zacarias",
    "malaquias": "malaquias",
    "mateus": "mateus",
    "marcos": "marcos",
    "lucas": "lucas",
    "joao": "joao",
    "joão": "joao",
    "atos": "atos",
    "romanos": "romanos",
    "1corintios": "icorintios",
    "1coríntios": "icorintios",
    "2corintios": "iicorintios",
    "2coríntios": "iicorintios",
    "galatas": "galatas",
    "gálatas": "galatas",
    "efesios": "efesios",
    "efésios": "efesios",
    "filipenses": "filipenses",
    "colossenses": "colossenses",
    "1tessalonicenses": "itessalonicenses",
    "2tessalonicenses": "iitessalonicenses",
    "1timoteo": "itimoteo",
    "1timóteo": "itimoteo",
    "2timoteo": "iitimoteo",
    "2timóteo": "iitimoteo",
    "tito": "tito",
    "filemom": "filemom",
    "hebreus": "hebreus",
    "tiago": "tiago",
    "1pedro": "ipedro",
    "2pedro": "iipedro",
    "1joao": "ijoao",
    "1joão": "ijoao",
    "2joao": "iijoao",
    "2joão": "iijoao",
    "3joao": "iiijoao",
    "3joão": "iiijoao",
    "judas": "judas",
    "apocalipse": "apocalipse"
}

# Solicitar ao usuário o livro, capítulo e versículo(s), como "Êxodo 20:11"
entrada = input("Digite o livro, capítulo e versículo(s) (ex: Êxodo 20:11): ").strip().lower()

# Separar o livro, capítulo e versículo
try:
    livro, capitulo_verso = entrada.split()
    capitulo, versiculo_str = capitulo_verso.split(':')
    versiculo = int(versiculo_str)
except ValueError:
    print("Erro: Certifique-se de inserir o livro, capítulo e versículo.")
    exit(1)

# Converter o nome do livro para a versão simplificada na URL
livro_simplificado = bible_books.get(livro)

if livro_simplificado:
    url = f"https://biblia-a-mensagem.vercel.app/{livro_simplificado}/{capitulo}"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        page_text = soup.get_text(separator=' ')

        # Remover todo o conteúdo anterior à palavra-chave "theme"
        keyword = "theme"
        keyword_idx = page_text.find(keyword)

        if keyword_idx != -1:
            page_text = page_text[keyword_idx + len(keyword):].strip()

        # Encontrar e remover a frase de rodapé
        unwanted_phrase = "Anterior Próximo"
        start_idx = page_text.find(unwanted_phrase)

        if start_idx != -1:
            page_text = page_text[:start_idx].strip()

        # Encontrar parágrafos numerados como "1-2", "3-5", etc.
        paragraphs = re.split(r'(?=\d+-\d+)', page_text)

        # Procurar o parágrafo que contém o versículo fornecido
        matching_paragraph = None
        for paragraph in paragraphs:
            match = re.match(r'(\d+)-(\d+)', paragraph)
            if match:
                start_verse, end_verse = map(int, match.groups())
                if start_verse <= versiculo <= end_verse:
                    matching_paragraph = paragraph
                    break

        if matching_paragraph:
            print(f"Conteúdo de {livro_simplificado.capitalize()} {capitulo}:\n")
            print(matching_paragraph.strip())
        else:
            print(f"O versículo {versiculo} não foi encontrado no capítulo {capitulo}.")
    else:
        print("Erro ao acessar a página. Verifique a URL ou a conexão.")
else:
    print("Livro não encontrado no dicionário.")
```
